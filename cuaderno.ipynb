### Muestreo sistemático

```python
def systematic_sampling(econdata,step):
    indexes = np.arange(0,len(econdata),step=step)
    systematic_sample = econdata.iloc[indexes]
    return systematic_sample
systematic_sample = systematic_sampling(econdata,3)
systematic_sample
```

### Estratificado

```python
econdata['estratificado'] = econdata['delegacion']+', '+econdata['tipo']
(econdata['estratificado'].value_counts()/len(econdata)).sort_values(ascending=False)
```

-- Los datos deberían ser, de forma ideal:
La porporción es la siguiente:
1. Hoteles en Cuauhtémoc: $0.5$
2. Museos en Cuauhtémoc: $0.2$
3. Hoteles en Venustiano Carranza: $0.1$
4. Mercados en Cuauhtémoc: $0.1$
5. Mercados en Veustiano Carranza: $0.1$

```python
def data_estratificada(econdata, nombres_columnas_estrat, valores_estrat, prop_estrat, random_state=None):
    
    df_estrat = pd.DataFrame(columns = econdata.columns) # Creamos un data frame vacío con los nombres de las columnas de econdata

    pos = -1
    for i in range(len(valores_estrat)): # iteración sobre los valores estratificados
        pos += 1
        if pos == len(valores_estrat) - 1: 
            ratio_len = len(econdata) - len(df_estrat) # si es la iteración final calcula el número de valores de salida tenga el mismo número de filas que de entrada
        else:
            ratio_len = int(len(econdata) * prop_estrat[i]) # calcula el número de filas según la proporción deseada

        df_filtrado = econdata[econdata[nombres_columnas_estrat] ==valores_estrat[i]] # filtra los datos de origen según los valores seleccionados en la estratificación de datos
        df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state) # haz un sample de los datos filtrados usando la ratio que hemos calculado
        
        df_estrat = pd.concat([df_estrat, df_temp]) # junta las tablas de sample con la estratificada para producir el resultado final
        
    return df_estrat # Return the stratified, re-sampled data   
```

```python
valores_estrat = ['Cuautémoc, Hotel', 'Cuautémoc, Museo', 'Venustiano Carranza, Hotel', 'Cuauhtémoc, Mercado','Venustiano Carranza, Mercado']
prop_estrat = [0.5, 0.2, 0.1, 0.1,0.1]
df_estrat = data_estratificada(econdata, 'estratificado', valores_estrat, prop_estrat, random_state=42)
df_estrat
```

### Intervalos de confianza
* Un par o varios pares de números entre los cuales se estima que estará cierto valor desconocido respecto de un parámetro poblacional con un determinado nivel de confianza.
* Son simétricos respecto a la media.

### Nivel de significación
* EL nivel de significación o alfa es el nivel límite para juzgar si un resultado es o no estadísticamente significativo.
* Si el valor de significación es menor que el nivel de significación, el resultado es estadísticamente significativo.

```python
import scipy.stats as st
```

```python
"""
El ejercicio es días de vacaciones con 0 siendo no tiene y 10 siendo
tiene vacaciones.
0 siendo no gasta nada y 10 siendo lo gastó todo el dinero
"""
np.random.seed(20) #Semilla aleatoria
var1 = np.random.randint(0,10,10) #La variable aleatoria de Días de vacaciones
var2 = var1+np.random.normal(0,1,10) #Dinero restante
```

```python
var1
```

```python
var2
```

```python
sns.regplot(x=var1,y=var2, ci=80)
```

```python
tuple = list(zip(var1,var2))
tuple
```

```python
tabla = pd.DataFrame(tuple,columns=['Dias_vacaciones','Dinero_gastado'])
tabla.sort_values(by='Dinero_gastado',ascending=False)
```

```python
# al 95%
st.t.interval(confidence = 0.95,df=len(tabla)-1,
             loc=np.mean(tabla),scale=st.sem(tabla))
```

#### Debido a que la primera variable son días, concerteza con los datos obtenidos, las personas se van 3 días. Entre tanto, con respecto al dinero gastado, estaría entre el 69% y el 70%.


#### Tambien se puede usar la funcion st.norm.interval para cuando se tiene una cantidad de muestras (n) mayor o igual a 30, esto debido a que con esta cantidad de muestras se puede asumir que los datos están normalmente distribuidos (teorema del limite central)

#### En resumen:

* Si n<30 usamos st.t.interval

* Si n>=30 usamos st.norm.interval

## Tipos de hipótesis

```python
from scipy.stats import f_oneway
from scipy.stats import ttest_ind
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
iris = pd.read_csv(url, names=names)
iris.head()
```

```python
iris.shape
```

### Distribución t-student

Sepal-lenght, sepal-width

Captura de pantalla 2022-03-31 a la(s) 10.14.41.png

```python
se_length = iris['sepal-length'].std()/np.sqrt(10)
se_width = iris['sepal-width'].std()/np.sqrt(10)
```

Captura de pantalla 2022-03-31 a la(s) 10.14.46.png

```python
sed = np.sqrt(se_length**2+se_width**2)
```

Captura de pantalla 2022-03-31 a la(s) 10.19.24.pngCaptura de pantalla 2022-03-31 a la(s) 10.17.14.png

```python
t = (iris['sepal-length'].mean()-iris['sepal-width'].mean())/sed
print('El cálculo de la t-student correspondería entonces a las medias',
      '\ndel length - el width porque así lo propusimos en el comienzo',
     f'\ny eso daría como resultado tras el uso del SED un valor t de: {t}')
```

```python
#Usando el ttest_ind:
p = ttest_ind(iris['sepal-length'],iris['sepal-width'])
p
```

#### Conclusiones del resultado anterior
El resultado de la prueba ttest_ind indica que el valor crítico, pvalue es de 3.98e-112 así que el valor absoluto de t que nos dió 9.43 es más grande que el p value

$t>pvalue$

Por lo cual podemos afirmar que las distribuciones son diferentes, rechazamos la hipóstesis de que las distribuciones de ambas fueran muy similares y nos vamos por una H1.

### Bootstrapping
